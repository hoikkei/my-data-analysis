- ## Single Stock Trading Strategy 1: Buy at the beginning of the quarter and sell at the end of the quarter
Trading period: 20230101-20240331          Initial capital: 100,000 USD

```python
import pandas as pd

# Load the data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

capital = 100000  # Load the data
trade_count = 0  # Number of trades
shares = 0  # Current amount of shares held

# Process data for each quarter
quarterly_data = df.resample('Q')

for name, group in quarterly_data:
    # Get the first and last day of each quarter
    first_day = group.iloc[0].name
    last_day = group.iloc[-1].name

    # Purchase
    if shares == 0:
        shares = capital // group.loc[first_day, 'close']
        capital -= shares * group.loc[first_day, 'close']
        trade_count += 1

    # Sell
    if shares > 0:
        capital += shares * group.loc[last_day, 'close']
        shares = 0
        trade_count += 1

# Calculate final returns
final_returns = capital - 100000

# Print results
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns:.2f} dollars")
print(f"The rate of return of the portfolio is: {final_returns * 100 / 100000:.2f}%")
```

- ## Single Stock Trading Strategy 2: Buy at the beginning of the month and sell at the end of the month
Trading period: 20230101-20240331 Initial capital: 100,000 USD

```python
import pandas as pd

# Load data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

capital = 100000  # initial capital
trade_count = 0  # number of trades
portfolio = []  # store portfolio value after each trade

# Group by month
monthly_data = df.resample('M')

for name, group in monthly_data:
    # Get first and last day of each month
    first_day = group.iloc[0].name
    last_day = group.iloc[-1].name

    # Get close price of first and last day
    buy_price = group.loc[first_day, 'close']
    sell_price = group.loc[last_day, 'close']

    # Calculate shares and update capital
    shares = capital / buy_price
    capital = shares * sell_price

    # Count trade
    trade_count += 1

    # Store portfolio value
    portfolio.append(capital)

print(f"After {trade_count} trades, the total return of the portfolio is: {capital - 100000:.2f} dollars")

# Calculate return rate
return_rate = (portfolio[-1] - 100000) / 100000
print(f"The return rate of the portfolio is: {return_rate * 100:.2f}%")
```

- ## Single Stock Trading Strategy 3: Double Moving Average Strategy
A moving average is a line that calculates the average closing price over a certain period. Common moving averages include 5-day, 10-day, 30-day, 60-day, 120-day, and 240-day. The formula for a moving average is as follows:MA (Moving Average) = (C1+C2+C3+...+Cn) / N C: The closing price for a certain day N: The period (number of days) for the moving average.

In stock price analysis terminology, golden cross and death cross are defined as follows: There are two lines of short-term and long-term moving averages for a certain indicator. When the short-term moving average crosses the long-term moving average from above, it is called a "golden cross", and when it crosses from below, it is called a "death cross".

Generally, a golden cross is considered a buying opportunity, and a death cross is considered a selling opportunity. However, a golden cross and a death cross are just one of many indicators, and they need to be used in combination with many other indicators to make the timing of trades more accurate.

Short-term trading: 5-day moving average and 10-day moving average 
Medium-term trading: 30-day moving average and 60-day moving average

```python
# Calculate the moving average
df['5_ma'] = df['close'].rolling(window=5).mean()
df['10_ma'] = df['close'].rolling(window=10).mean()
df['30_ma'] = df['close'].rolling(window=30).mean()
df['60_ma'] = df['close'].rolling(window=60).mean()
```

```python
# Create short-term and medium-term trading signals
df['short_signal'] = 0
df.loc[df['5_ma'] > df['10_ma'], 'short_signal'] = 1  # Short-term golden cross  
df.loc[df['5_ma'] < df['10_ma'], 'short_signal'] = -1  # Short-term death cross

df['mid_signal'] = 0
df.loc[df['30_ma'] > df['60_ma'], 'mid_signal'] = 1  # Medium-term golden cross
df.loc[df['30_ma'] < df['60_ma'], 'mid_signal'] = -1  # Medium-term death cross
```

```python
# Calculate the strategy return
initial_cash = 100000
df['short_holdings'] = df['short_signal'].shift(1)  # Short-term holdings
df['mid_holdings'] = df['mid_signal'].shift(1)  # Medium-term holdings
df['short_cash'] = initial_cash  # Initial cash
df['mid_cash'] = initial_cash
df['short_stock_value'] = df['short_holdings'] * df['close']  # Short-term position value
df['mid_stock_value'] = df['mid_holdings'] * df['close']  # Medium-term position value
df['short_cash_flow'] = df['short_holdings'].diff() * df['close']  # Short-term trading cash flow
df['mid_cash_flow'] = df['mid_holdings'].diff() * df['close']  # Medium-term trading cash flow
df['short_cash'] = df['short_cash'] - df['short_cash_flow'].fillna(0).cumsum() # Calculate the short-term cash curve
df['mid_cash'] = df['mid_cash'] - df['mid_cash_flow'].fillna(0).cumsum() # Calculate the medium-term cash
df['short_total_equity'] = df['short_cash'] + df['short_stock_value']  # Short-term total equity curve
df['mid_total_equity'] = df['mid_cash'] + df['mid_stock_value']  # Medium-term total equity curve
```

```python
# Calculate the number of trades
short_trade_count = (df['short_signal'].diff().abs() != 0).sum()
mid_trade_count = (df['mid_signal'].diff().abs() != 0).sum()

# Calculate the return
short_return = df['short_total_equity'].iloc[-1] - initial_cash
mid_return = df['mid_total_equity'].iloc[-1] - initial_cash

# Calculate the return rate
short_return_rate = short_return / initial_cash
mid_return_rate = mid_return / initial_cash

print(f"After {short_trade_count} short-term trades, the total return of the portfolio is: {short_return:.2f} dollars")
print(f"The return rate of the short-term portfolio is: {short_return_rate * 100:.2f}%")

print(f"After {mid_trade_count} mid-term trades, the total return of the portfolio is: {mid_return:.2f} dollars")
print(f"The return rate of the mid-term portfolio is: {mid_return_rate * 100:.2f}%")
```

- ## Single Stock Trading Strategy 4: Bollinger Bands Strategy

Bollinger Bands consist of three lines, with the top and bottom two lines representing the resistance and support lines for the price, respectively. The middle line represents the moving average of the price. The formulas are as follows:

Middle Line = 20-day Moving AverageUpper Line = 20-day Moving Average + N Standard Deviations of 20-day TerminalLower Line = 20-day Moving Average - N Standard Deviations of 20-day Terminal

Bollinger Bands Strategy: If the price breaks through the resistance line, sell all, and if it falls below the support line, buy all. One of the challenges in the study of the Bollinger Bands strategy is the choice of the value of N and the setting of the band width. 

This strategy is based on the premise that when the price exceeds the upper and lower bands, it tends to return to the average. However, this premise may not always hold true. Under certain market conditions such as a trending market or a sideways market, the effectiveness of this strategy may vary greatly.

```python
import pandas as pd

# Load the data
data = pd.read_csv('./NVDA_stock_data_since_2023.csv', index_col='date', parse_dates=True)

# Calculate Bollinger Bands
data['MA20'] = data['close'].rolling(window=20).mean()
data['20dSTD'] = data['close'].rolling(window=20).std() 
data['Upper'] = data['MA20'] + (data['20dSTD'] * 2)
data['Lower'] = data['MA20'] - (data['20dSTD'] * 2)

# Set initial capital and number of shares
capital = 100000
shares = 0

# Initialize trade count
trade_count = 0

# Traverse stock data and trade according to Bollinger Band strategy
for i in range(20, len(data)):
    # Today's closing price
    close = data['close'][i]
    
    # If closing price is lower than lower Bollinger Band, buy stock
    if close < data['Lower'][i]:
        shares = capital / close  # Number of shares bought
        capital = 0  # Remaining capital after buying stocks is 0
        trade_count += 1  # Increase trade count
        
    # If closing price is higher than upper Bollinger Band and shares > 0, sell stock
    elif close > data['Upper'][i] and shares > 0:
        capital = shares * close  # Capital after selling stocks
        shares = 0  # Number of shares after selling is 0
        trade_count += 1  # Increase trade count

# Calculate final returns
if shares > 0:
    capital = shares * data['close'][-1]

# Calculate return rate
return_rate = (capital - 100000) / 100000

# Print results according to your format
print(f"After {trade_count} trades, the total return of the portfolio is: {capital - 100000:.2f} dollars")
print(f"The return rate of the portfolio is: {return_rate * 100:.2f}%")
```

MACD is a momentum oscillation indicator calculated from the difference of two Exponential Moving Averages (EMAs). 
You buy when the MACD line crosses the signal line from below, and sell when it crosses from above.

If the CCI value is greater than 0, it indicates that the stock price might have exceeded the average price within N periods, suggesting a potential overbuying situation. 

If the CCI value is less than 0, it indicates that the stock price might be below the average price within N periods, suggesting a potential overselling situation.

In this strategy, you buy stocks when the CCI value is positive and sell stocks when the CCI value is negative. This is because a positive CCI value indicates that the stock price has exceeded the average price within N periods, suggesting a potential upward price trend.
```python
# Load data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Calculate Typical Price
df['TP'] = (df['high'] + df['low'] + df['close']) / 3

# Calculate CCI
N = 14
df['TP_MA'] = df['TP'].rolling(window=N).mean()
df['TP_MD'] = df['TP'].rolling(window=N).apply(lambda x: np.mean(np.abs(x - np.mean(x))))
df['CCI'] = (df['TP'] - df['TP_MA']) / (0.015 * df['TP_MD'])

# Create a column for holding the trading signals
df['Signal'] = 0
df.loc[df['CCI'] > 0, 'Signal'] = 1  # Buy when CCI is positive
df.loc[df['CCI'] < 0, 'Signal'] = -1  # Sell when CCI is negative

# Execute the trading strategy
capital = 100000  # initial capital
shares = 0  # number of shares held
trade_count = 0  # number of trades

for i in range(len(df)):
    # Buy
    if df['Signal'][i] > 0 and shares == 0:
        shares = capital // df['close'][i]  # buy as many shares as possible
        capital -= shares * df['close'][i]  # update capital
        trade_count += 1  # increment trade counter
    # Sell
    elif df['Signal'][i] < 0 and shares > 0:
        capital += shares * df['close'][i]  # sell all shares
        shares = 0  # update shares
        trade_count += 1  # increment trade counter

# Calculate final returns
final_returns = (capital + shares * df['close'][-1] - 100000) / 100000

# print the result
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns * 100000:.2f} dollars")
print(f"The return rate of the portfolio is: {final_returns * 100:.2f}%")
```

- ## Single Stock Trading Strategy 19: TEMA Grid Strategy

The Triple Exponential Moving Average (TEMA) strategy is mainly used to identify trends, rebounds, and potential support and resistance lines.
Like other moving average strategies, consider a buy order when the indicator is upwards and a sell order when it is downwards. In this strategy, you first calculate the TEMA, and generate buying and selling signals based on its changes. If the TEMA is rising, you place a buy order, and if it is falling, you place a sell order. Finally, calculate the daily returns and cumulative returns.

However, TEMA has the potential to overly smooth price data, especially when dealing with high volatility tech stocks. This smoothing can result in delayed or distorted signals. This is because TEMA may mistakenly recognize short-term price fluctuations as part of the trend, and it may not effectively filter out noise.

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Calculate TEMA
EMA1 = df['close'].ewm(span=5).mean()
EMA2 = EMA1.ewm(span=5).mean()
EMA3 = EMA2.ewm(span=5).mean()
df['TEMA'] = 3*EMA1 - 3*EMA2 + EMA3

# Create a column for holding the trading signals
df['Signal'] = 0

# Generate trading signals
df.loc[df['TEMA'] > df['TEMA'].shift(), 'Signal'] = 1
df.loc[df['TEMA'] < df['TEMA'].shift(), 'Signal'] = -1

# Calculate daily returns
df['Return'] = df['close'].pct_change() * df['Signal'].shift()

# Calculate total trades
trade_count = (df['Signal'].diff() != 0).sum()

# Calculate final returns
final_returns = df['Return'].sum()

# Print the result
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns * 100000:.2f} dollars")
print(f"The return rate of the portfolio is: {final_returns * 100:.2f}%")
```

Single Stock Trading Strategy 20: KDJ StrategyThe KDJ Stochastic Strategy is based on the random indicator, used to identify the overbought and oversold conditions in the market. 

In this strategy, you first calculate the K value, D value, and J value and then generate buying and selling signals based on the golden cross and dead cross of the K line and D line.

However, under high volatility market conditions, KDJ Stochastics may generate excessive buying and selling signals, especially when the market is sideways or there is no clear trend.
```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Calculate KDJ
low_list = df['low'].rolling(9, min_periods=1).min()
high_list = df['high'].rolling(9, min_periods=1).max()
rsv = (df['close'] - low_list) / (high_list - low_list) * 100

df['K'] = pd.DataFrame(rsv).ewm(com=2).mean()
df['D'] = df['K'].ewm(com=2).mean()
df['J'] = 3 * df['K'] - 2 * df['D']

# Create a column for holding the trading signals
df['Signal'] = 0

# Generate trading signals
df.loc[df['K'] > df['D'], 'Signal'] = 1
df.loc[df['K'] < df['D'], 'Signal'] = -1

# Calculate daily returns
df['Return'] = df['close'].pct_change() * df['Signal'].shift()

# Calculate total trades
trade_count = (df['Signal'].diff() != 0).sum()

# Calculate final returns
final_returns = df['Return'].sum()

# Print the result
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns * 100000:.2f} dollars")
print(f"The return rate of the portfolio is: {final_returns * 100:.2f}%")
```

- ### Quarterly Buy/Sell Strategy based on EMA, BOP

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Calculate the short-term and long-term EMA
short_period = 12
long_period = 26
df['ShortEMA'] = df['close'].ewm(span=short_period, adjust=False).mean()
df['LongEMA'] = df['close'].ewm(span=long_period, adjust=False).mean()

# Calculate BOP
df['BOP'] = (df['close'] - df['open']) / (df['high'] - df['low'])

# Create a column to store the trading signals
df['Signal_EMA'] = 0
df.loc[df['ShortEMA'] > df['LongEMA'], 'Signal_EMA'] = 1  # Buy when short-term EMA is greater than long-term EMA
df.loc[df['ShortEMA'] < df['LongEMA'], 'Signal_EMA'] = -1  # Sell when short-term EMA is less than long-term EMA

df['Signal_BOP'] = 0
df.loc[df['BOP'] > 0, 'Signal_BOP'] = 1  # Buy when BOP is positive
df.loc[df['BOP'] < 0, 'Signal_BOP'] = -1  # Sell when BOP is negative

# Process the data for each quarter
quarterly_data = df.resample('Q')  # The change is here, from 'M' to 'Q'

capital = 100000  # Initial capital
shares = 0  # Number of shares held
trade_count = 0  # Number of trades

for name, group in quarterly_data:
    # Get the first and last day of each quarter
    first_day = group.iloc[0].name
    last_day = group.iloc[-1].name

    # If the signals from EMA and BOP are the same, execute the trade
    if (group.loc[first_day, 'Signal_EMA'] == group.loc[first_day, 'Signal_BOP']) and (group.loc[first_day, 'Signal_EMA'] == 1):
        # Buy
        shares = capital // group.loc[first_day, 'close']  # Buy as many shares as possible
        capital -= shares * group.loc[first_day, 'close']  # Update capital
        trade_count += 1  # Count number of trades

    elif (group.loc[last_day, 'Signal_EMA'] == group.loc[last_day, 'Signal_BOP']) and (group.loc[last_day, 'Signal_EMA'] == -1) and shares > 0:
        # Sell
       capital += shares * group.loc[last_day, 'close']  # Sell all shares
       shares = 0  # Update number of shares
       trade_count += 1  # Count number of trades

# Calculate the final returns
final_returns = capital + shares * df['close'][-1] - 100000

# Print the results
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns:.2f} dollars")
print(f"The rate of return of the portfolio is: {final_returns * 100 / 100000:.2f}%")
```

- ## Short-term Trading + EMA Strategy
After two trades, the total return of the portfolio is -4989.60 dollars. The return rate of the portfolio is -4.99%. However, it did not match the profitability of the CCI and MACD indicators.

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('./NVDA_stock_data_since_2023.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Select data only from 2024-03-07 to 2024-03-28
start_date = '2024-03-07'
end_date = '2024-03-28'
df = df.loc[start_date:end_date]

# Calculate short-term and long-term EMA
short_period = 12
long_period = 26
df['ShortEMA'] = df['close'].ewm(span=short_period, adjust=False).mean()
df['LongEMA'] = df['close'].ewm(span=long_period, adjust=False).mean()

# Create a column for holding the trading signals
df['Signal'] = 0
df.loc[df['ShortEMA'] > df['LongEMA'], 'Signal'] = 1  # Buy when short-term EMA is above long-term EMA
df.loc[df['ShortEMA'] < df['LongEMA'], 'Signal'] = -1  # Sell when short-term EMA is below long-term EMA

# Execute the trading strategy
capital = 100000  # initial capital
shares = 0  # number of shares held
trade_count = 0  # number of trades

for i in range(len(df)):
    # Buy
    if df['Signal'][i] > 0 and shares == 0:
        shares = capital // df['close'][i]  # buy as many shares as possible
        capital -= shares * df['close'][i]  # update capital
        trade_count += 1  # increment trade counter
    # Sell
    elif df['Signal'][i] < 0 and shares > 0:
        capital += shares * df['close'][i]  # sell all shares
        shares = 0  # update shares
        trade_count += 1  # increment trade counter

# Calculate final returns
final_returns = (capital + shares * df['close'][-1] - 100000) / 100000

# print the result
print(f"After {trade_count} trades, the total return of the portfolio is: {final_returns * 100000:.2f} dollars")
print(f"The return rate of the portfolio is: {final_returns * 100:.2f}%")
```
